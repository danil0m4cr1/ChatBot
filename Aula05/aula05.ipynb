```python
from google.colab import userdata
chat_api = userdata.get("api_secret2")
```

```python
import gradio as gr
from groq import Groq
```

```python
# cria um cliente para receber o modelo LLM
client = Groq(
    api_key=chat_api
)

# cria uma função chamada assistente agricultura
def assistente_agricultura(user_prompt:str) ->str:
    try:
      if user_prompt.strip() == "15":
        return "Encerrando o Chatbot"
      completion = client.chat.completions.create(
          model = "llama-3.1-8b-instant",
          messages=[
              {"role":"system","content":"Você é um especialista em química e física"},
              {"role":"user","content":user_prompt}
              ],
          temperature = 0,
          max_tokens=1024,
          top_p=1,
          stream = False
      )

      return completion.choices[0].message.content

    except Exception as e:
      print("\n Erro no especialista em química e física")
      return f"Erro {e}"

iface = gr.Interface(
    fn = assistente_agricultura,
    inputs = gr.Textbox(lines=10, placeholder="Digite sua pergunta sobre química e física"),
    outputs=gr.Textbox(lines=10,label="Resposta"),
    title="Especialista em química e física",
    description="Digite sua pergunta sobre química e física e receba respostas da LLM especialista",
    live = False
)

if __name__ == "__main__":
  iface.launch(debug=True, show_error=True)
```
